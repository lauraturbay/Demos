## Import libraries

````python
## import libraries
import pandas as pd
````

## Upload Olympics datasets to this notebook.

### Import ```olympics-summer.csv``` and store in a dataframe called ```dfs``` (<a href="https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-summer.csv">link</a>).


### Import ```olympics-winter.csv``` and store in a dataframe called ```dfw``` (<a href="https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-winter.csv">link</a>).

````python
dfs = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-summer.csv")
dfw = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/olympics-winter.csv")
````

````python
## subsetting dataframe into smaller, targeted datframes 
## specific columns
````

````python
## filtered data to just three required columns

df_sub = dfw[["Year","City","Team"]]
df_sub
````

````python
## .filter() to filter down a large datframe
dfw.filter(["Year","City","Team"])
````

````python
dfw.filter(like="eight")
````

````python
## sort by athletes names A-Z. Sorting.

dfw.sort_values(by="Athlete Name", ascending= True)
````

````python
##sort by city and year. Use another list, to do more than one

dfw.sort_values(by = ["City", "Year"], ascending = False)
````

````python
dfs["Season"] = "Summer"
dfs.info()
````

````python
dfw["Season"] = "Winter"
dfw.head()
````

````python
list(dfs.columns) == list(dfw.columns)
````

````python
df = pd.concat([dfw, dfs], ignore_index = True)
df
````

````python
df.info()
````

````python
df["Medal"].unique()
````

````python
df["Sex"].unique()
````

````python
## Save them as categories so the search for a unique indicator is faster
````

````python
df["Sex"] = df["Sex"].astype("category")
df.info()
````

````python
df[["Medal", "Season"]] = df[["Medal", "Season"]].astype("category")
df.info()
````

````python
## Find specific type of data. df that only holds Sochi

df_sochi = df.query('City == "Sochi"')
df_sochi
````

````python
df.query('Medal == "Gold"')
````

````python
## df holds oonly data ass w Jeremy Abbott. USE TICK MARK

df_abbott = df.query('`Athlete Name` == "Jeremy Abbott"')
df_abbott
````

````python
df_1984 = df.query('Year == 1984')
df_1984
````

````python
##Return all the competitions between 1920 and 1950

df.query('1920 <= Year <= 1950')
````

````python
## find tennis and women only between 1972 and as recent as possible, data frame.

df.query('(Year > 1972) and (Sport == "Tennis") and (Sex == "F")')
````

````python
df_gold = df.query("Medal == 'Gold'")
df_gold
````

````python
## 10 countries with most gold

df_goldtop = df_gold.value_counts("Team").to_frame("Number Gold")

df_goldtop
````

````python
df_goldtop.head(10)
````

````python
## return as %

df_goldtop = (df_gold.value_counts("Team", normalize = True)*100).to_frame("PCT Gold").head(10)
df_goldtop
````
